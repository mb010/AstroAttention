## Select Model and Data

```python
paths=[
    '0709Exp1-range_norm-MiraBestNOHYBRIDAdam-AGRadGalNetft-5e-05'
]
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
nets = []
evaluations = []

for p in paths:
    model_name = path_to_model(p)
    nets.append(load_net(model_name, device))
```

# Count Incorrect Values Across Data Augmented Data Set (1000 Augmentations)

```python
net = nets[0]
false_count = np.zeros((153))

repetitions = 1000
for r in range(repetitions):
    data, out_date = data_call('MiraBest', dataloader=True)
    # Create set of sorted batch for each repetition
    count = 0
    for d in data:
        batch = d[0]
        labels= d[1].numpy()
        labels = np.where(labels<5,0,1)
        predictions = net(batch.to(device)).detach().cpu().numpy().argmax(axis=1)
        false = np.where(labels!=predictions,1,0)
        if labels.shape[0]!=16:
            false_count[count:count+labels.shape[0]] += false[:-4]
        else:
            false_count[count:count+labels.shape[0]] += false
        count+=labels.shape[0]
```

```python
data, out_data = data_call('MiraBest', dataloader=False)
print(out_data.data.squeeze().shape)
sources = out_data.data.squeeze()
```
